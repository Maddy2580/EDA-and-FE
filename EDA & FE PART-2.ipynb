```python
##which category having largetest number of installation.

df.groupby(['Category'])['Installs']
```

The code you provided groups a DataFrame called df by the "Category" column, and then selects the "Installs" column from each group. Here's a breakdown of what each part of the code does:

df: specifies the name of the DataFrame that is being grouped.
.groupby(['Category']): groups the DataFrame by the unique values in the "Category" column.
['Installs']: selects the "Installs" column from each group.
So, after executing this code, you will have a Pandas GroupBy object that contains groups of rows from df that have the same value in the "Category" column. For each group, the "Installs" column will be selected. This allows you to perform aggregate functions on the "Installs" column for each group, such as finding the sum, mean, or maximum value of installs for each category.






```python
df.groupby(['Category'])['Installs'].sum().sort_values(ascending=False).reset_index()
```

# The code you provided creates a new DataFrame called df_cat_installs that contains the sum of the "Installs" column for each unique value in the "Category" column of a DataFrame called df. The resulting DataFrame is sorted by the sum of installs in descending order. Here's a breakdown of what each part of the code does:

# df: specifies the name of the DataFrame that is being grouped.

# .groupby(['Category']): groups the DataFrame by the unique values in the "Category" column.

# ['Installs']: selects the "Installs" column from each group.

# .sum(): computes the sum of the "Installs" column for each group.


# .sort_values(ascending=False): sorts the resulting Series of sums in descending order, so that the categories with the highest number of installs appear first.


# .reset_index(): resets the index of the resulting DataFrame so that the "Category" column is no longer the index, but rather a regular column with a numerical index.


# So, after executing this code, df_cat_installs will be a DataFrame that has two columns: "Category" and "Installs". The "Category" column will contain the unique values from the "Category" column of df, and the "Installs" column will contain the sum of installs for each category, sorted in descending order.

```python
df_cat_installs=df.groupby(['Category'])['Installs'].sum().sort_values(ascending=False
).reset_index()
```

```python
### What are the Top 5 most installed Apps in Each popular Categories ??

df.groupby(['Category','App'])['Installs'].sum().reset_index()
```

```python
dfa=df.groupby(['Category','App'])['Installs'].sum().reset_index()
```

```python
dfa=dfa.sort_values('Installs', ascending=False)
```

```python
dfa[dfa.Category=='GAME'][:5]
```

```python
dfa[dfa.Category=='PRODUCTIVITY'][:5]
```

```python
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")
plt.figure(figsize=(40,30))
for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,y= 'Installs' ,x='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

# So, after executing this code, you can use the variables i and app within the loop to perform operations on each element in the apps list. The enumerate() function is useful for accessing the index of an element in a list or array when you need to perform operations that require that information, such as updating another list or selecting an element from a corresponding list.
